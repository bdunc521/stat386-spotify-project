##### Libraries and data clean up

```python
import pandas as pd
import ast
df = pd.read_csv('data.csv')

#Turn Artists into two seperate artist
df['artists'] = df['artists'].apply(ast.literal_eval)

# One row per artist
df = df.explode('artists')
```

##### Taylor Swift

```python
tswift = df[df['artists'].str.lower() == 'taylor swift'].reset_index(drop = True)
tswift.head()
```

##### Drake

```python
drake = df[df['artists'].str.lower() == 'drake'].reset_index(drop = True)
drake.head()
```

```python
df = pd.concat([kanye, drake, bunny, bruno, kperry, tswift, hstyles, morgan, tame, jackson], axis = 0).reset_index(drop=True)

df = df[df['duration_ms'] > 53000]
df['duration_ms'] = df['duration_ms']/1000

df = df.rename(columns = {'duration_ms' : 'duration_s'})

df['Year_dt'] = pd.to_datetime(df['year'], format='%Y')
```

##### Save to a CSV

```python
df.to_csv('app_artists.csv')
```

```python
df = df[df['artists']=='Kanye West']
durat_avg = df.groupby('year')['duration_s'].mean().reset_index()

import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.scatter(durat_avg['year'], durat_avg['duration_s'] , color='red')
ax.set_title('Average Yearly Song Length')
ax.set_xlabel('Year')
ax.set_ylabel('Length (s)')
ax.patch.set_alpha(0)

```
